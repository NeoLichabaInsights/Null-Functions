```sql
-- Create a table with the datatypes and constraints
CREATE TABLE products (
    product_id INT,
    name VARCHAR(50),
    price DECIMAL(10,2),
    discount DECIMAL(10,2)
);

-- Insert values into the table
INSERT INTO products (product_id, name, price, discount)
VALUES
(501, 'keyboard', 25, NULL),
(502, 'mouse', 15, 5),
(503, 'monitor', 100, NULL);

-- Show the contents of the table
SELECT * 
FROM products;

-- Q7. Show product name, price and final price after discount (assume 0 if discount is NULL)
-- Expected Columns: product_id, name, price, final_price
-- Table to yield value of the final price after discount has been factored in

SELECT product_id, 
       name,
       price,
       price - IFNULL(discount,0) AS final_price
FROM products;
```
